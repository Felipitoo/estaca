#### Interact 

Si trabaja con el kernel de python podrá utilizar "interact" que le permite hacer que su código sea interactivo, esto quedará más claro con el siguiente ejemplo. Interact puede incluir mas elementos que los que se presentan en este ejemplo.

````python
from ipywidgets import interact 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline 

def plot_serie(x):
    x_axis=np.arange(1,x,0.5)
    values = np.arange(1,x,0.5)**2
    values2 = np.arange(1,x,0.5)**3
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.plot(x_axis,values,label=r'$x^2$')
    plt.plot(x_axis,values2,label=r'$x^3$')
    plt.legend()

interact(plot_serie,x=(2,20))
````

## 3. Semilla 

A la hora de trabajar con números o procesos aleatorios a veces es importante poder replicar lo que usted realiza (pensando en un trabajo que se mostrará a la comunidad científica), para esto se tiene que, para generar la misma secuencia de números aleatorios, se debe pasar una *semilla* al generador de éstos. 

**R**
```R
set.seed(numero)
```

**Python**
```python
numpy.random.seed(numero) #libreria numpy 
random.seed(numero) #libreria random
#O de otra libreria que pueda utilizar para generar números aleatorios
```

Si desea puede aplicar esto en su laboratorio para que se pueda reproducir exactamente lo mismo que usted obtuvo.

````python
## Ejemplo R 

print("Obteniendo el mismo número siempre")
for(i in 1:10){
    set.seed(1) #Esto hará que saquemos el mismo número una y otra vez 
    resultado = sample(c(1:1000),1) #Sacaremos 1 número entre 1 y 1000
    print(resultado)
}

print("Obteniendo la misma secuencia ") 
for(j in 1:2){
    set.seed(1)
    print(" ")
    for(i in 1:10){
        resultado = sample(c(1:1000),1)
        print(resultado) #Aquí se obtiene la misma secuencia 
    }
}
````

````python
## Ejemplo Python 
import random 

print("Obteniendo el mismo número siempre \n")
for i in range(10):
    random.seed(1)
    resultado = random.randint(1,1000)
    print(resultado)

print("\n Obteniendo la misma secuencia ") 
for j in range(2):
    random.seed(1)
    print(" ")
    for i in range(10):
        resultado = random.randint(1,1000)
        print(resultado)
````
